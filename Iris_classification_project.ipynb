```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
%matplotlib inline
```

```python
columns = ['Sepal length','Sepal width','Petal length','Petal width','Class_labels']
df = pd.read_csv('Iris.csv',names = columns)
df.head()
```

```python
df.describe()
```

```python
for col in ['Sepal length', 'Sepal width', 'Petal length', 'Petal width']:
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
sns.pairplot(df, hue='Class_labels', vars=['Sepal length', 'Sepal width', 'Petal length', 'Petal width'])
```

```python
data = df.values
x = data[:,0:4]
y = data[:,4]
print(y)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2)
print(x_test)
```

Model 1: support Vector machine algorithm

```python
from sklearn.svm import SVC
from sklearn.impute import SimpleImputer
```

```python
imputer = SimpleImputer(strategy='mean')
```

```python
x_train = imputer.fit_transform(x_train)
x_test = imputer.transform(x_test)
```

```python
model_svc = SVC()
model_svc.fit(x_train,y_train)
```

```python
prediction1=model_svc.predict(x_test)
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,prediction1)*100)
for i in range(len(prediction1)):
  print(y_test[i],prediction1[i])
```

Model 2: Logistic Regression


```python
from sklearn.linear_model import LogisticRegression
model_LR=LogisticRegression()
model_LR.fit(x_train,y_train)
```

```python
prediction2 = model_LR.predict(x_test)
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,prediction2)*100)
for i in range(len(prediction1)):
  print(y_test[i],prediction1[i])
```
